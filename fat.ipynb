FIND THE GAUSSIAN DISTRIBUTION OF DEPENDENT VARIABLE WITH INDEPENDENT VARIABLE

```python
def displots(col):
    plt.figure(figsize=(10,7))
    sns.distplot(df['BodyFat'],color='#FF8787',hist=False,label='Body Fat')
    sns.distplot(df[col],color='#B1AFFF',hist=False,label=col)
    plt.legend()
    plt.show()
cols=list(df.columns)
for i in cols:
    print(f"The distribution plot for {i} feature is shown below:")
    displots(i)
    print('-'*100)
```

FIND THE PROBABILITY PLOT TO CHECK FOR OUTLIERS

```python
import scipy.stats as stats;

def drawplots(df,col):
    plt.figure(figsize=(12,6))
    plt.subplot(1,3,1)
    sns.histplot(df[col],color='#E0144C')
    
    plt.subplot(1,3,2)
    stats.probplot(df[col],dist='norm',plot=plt)
    
    plt.subplot(1,3,3)
    sns.boxplot(df[col],color='#905E96')
    plt.tight_layout()
    plt.show()
    
cols=list(df.columns)

for i in cols:
    print(f"the distribution plots for {i} feature is shown below:")
    drawplots(df,i)
    print("="*100)
```

CHECKING FOR OUTLIERS

```python
upperlimit=[]
lowerlimit=[]
for i in df.columns:
    upperlimit.append(df[i].mean()+(df[i].std())*4)
    lowerlimit.append(df[i].mean()-(df[i].std())*4)
```

```python
cols=list(df.columns)
j=0
for i in cols:
    temp=df.loc[(df[i]>upperlimit[j]) & (df[i]<lowerlimit[j])]
    j+=1
```

```python
temp
```

SINCE WE HAVE NONE OF THE VALUES SELELECTED THIS INDICATES THERE ARE NO OUTLIERS.GENERALLY OUTLIERS OCCUR IN SKEWED DISTRUBUTION.SINCE OUR FEATUERES FOLLOW GAUSSIAN DISTRIBUTION WE DONT FIND ANY OUTLIERS HERE.

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(),annot=True,cmap='plasma')
```

```python
def correlation(df,threshold):
    colcor=set()
    cormat=df.corr()
    for i in range(len(cormat)):
        for j in range(i):
            if abs(cormat.iloc[i][j])>threshold:
                 colname=cormat.columns[i]
                 colcor.add(colname)
    return colcor

ans=correlation(X,threshold=0.85)
ans
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV;
from sklearn import tree;
from sklearn.svm import SVR;
from sklearn.ensemble import RandomForestRegressor;
from sklearn.tree import DecisionTreeRegressor;
```

```python
x_train,x_test,y_train,y_test=train_test_split(train,Y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
reg=DecisionTreeRegressor()
reg.fit(x_train,y_train)
plt.figure(figsize=(20,12))
tree.plot_tree(reg,filled=True)
```

```python
path=reg.cost_complexity_pruning_path(x_train,y_train)
ccp_alpha=path.ccp_alphas
```

```python
alphalist=[]
for i in range(len(ccp_alpha)):
    reg=DecisionTreeRegressor(ccp_alpha=ccp_alpha[i])
    reg.fit(x_train,y_train)
    alphalist.append(reg)
```

```python
alphalist[:5]
```

```python
trainscore=[alphalist[i].score(x_train,y_train)for i in range(len(alphalist))]
testscore=[alphalist[i].score(x_test,y_test)for i in range(len(alphalist))]
plt.xlabel('Alpha')
plt.ylabel('Accuracy')
plt.plot(ccp_alpha,trainscore,label='train',color="#82CD47",drawstyle='steps-post')
plt.plot(ccp_alpha,testscore,label='test',color='#FF6464',drawstyle='steps-post')
plt.legend()
plt.show()
```

```python
score,reg.score(x_train,y_train)
```

```python
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
rf=RandomForestRegressor(n_estimators=100,ccp_alpha=1)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
score=r2_score(y_test,y_pred)
score,rf.score(x_train,y_train)
```
